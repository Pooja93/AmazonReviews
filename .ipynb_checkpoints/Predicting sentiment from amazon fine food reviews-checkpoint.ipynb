```python
import graphlab
food_reviews = graphlab.SFrame('Downloads/amazon-fine-foods/Reviews.csv')
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
food_reviews['word_count'] = graphlab.text_analytics.count_words(food_reviews['Summary'])
```

```python
quaker_cookie_reviews = food_reviews[food_reviews['ProductId']=='B007JFMH8M']
```

```python
food_reviews = food_reviews[food_reviews['Score'] != 3]
```

```python
food_reviews['sentiment'] = food_reviews['Score'] >=4
```

```python
food_reviews.head()
```

```python
train_data,test_data = food_reviews.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                     target='sentiment',
                                                     features=['word_count'],
                                                     validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```

```python
quaker_cookie_reviews = food_reviews[food_reviews['ProductId']=='B007JFMH8M']
```

```python
quaker_cookie_reviews['predicted_sentiment'] = sentiment_model.predict(quaker_cookie_reviews, output_type='probability')
```

```python
quaker_cookie_reviews = quaker_cookie_reviews.sort('predicted_sentiment', ascending=True)
```
